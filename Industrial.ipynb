# Distribuciones por categoria y macrodistrito - Cat industrial

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
direccion = './'
archivo = 'meses_total.xlsx'
fuente = direccion +'/'+ archivo
df = pd.read_excel(fuente)
display(df)
```

```python
df.dropna(inplace=True)
df.info()
```

# Filtrado por Categoria Industrial

```python
industrial = df[df['categoria']=='INDUSTRIAL']
industrial
```

# Filtrado por macrodistritos

# Centro

```python
D_Centro = 'C:/Users/Jose Ibanez/Datos/DATARMRV/'
F_Centro = 'ConsuIndustrial.xlsx'
Fuente_C = D_Centro+'/'+F_Centro
centro = pd.read_excel(Fuente_C, sheet_name='Centro')
centro.info()
```

```python
centro = centro.set_index("CLIENTE")
```

```python
industrial  = industrial.set_index("CLIENTE")
```

```python
centro_merge = pd.merge(centro, industrial, how='left',left_index=True,right_index=True)
centro_merge
```

```python
Macro_centro =centro_merge[['ene','feb','mar','abr','may','jun','jul','ago','sep','oct','nov','dic']].dropna()
```

```python
Macro_centro
```

```python
Macro_centro.to_csv('Centro_industrial_2023.csv')
Macro_centro.to_excel('Centro_industrial_2023.xlsx')
```
